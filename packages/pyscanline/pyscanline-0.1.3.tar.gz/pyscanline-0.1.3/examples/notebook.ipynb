```python
from pyscanline import scanline, scanline_gif, convert_frames_to_gif   
```

```python
out_img = scanline(
    image="alien_ripley.jpg",
    n_scanlines=90,
    scanline_col=("black", "darkslategrey", "#b4eeb4", "paleturquoise"),
    opacities=(1, 0.6, 0, 0, 0.6, 1),
    normalise=False,
    border_size=1,
    border_intensity=255,
    frame_size=10,
    vertical_filter="point",
    horizontal_filter="triangle",
    add_noise=False,
    noise_type="gaussian",
    print_details=False,
)
out_img.save("alien_ripley.scan.png")
out_img
```

```python
# Example usage
frames, durations = scanline_gif(
    gif="alien-1979.gif",
    width=762,
    height=456,
    scale=1,
    fps=10,
    delay=None,
    add_noise=True
)
convert_frames_to_gif(frames, "alien-1979.scan.gif")
```
